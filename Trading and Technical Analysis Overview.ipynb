Fibonacci Example - Monthly time frame of the S&P500 (Nasdaq top 500 companies on the stock market - total market capitalization)



The red arrows are how to find the fibonacci extension. Start from the low point of the wick of the candle, to the peak and then down to the low. This fibonacci extension is drawn from March 2009 to June 2010. Key .618 levels are respected into the future as highlighted with the yellow arrows. The first yellow arrow from the left shows that the .618 level acted as resistance until it was broken, to the up side, then fell back down for a month only to finally break in a seemingly constant growth period that broke past the 1.618 level only to act as support as highlighted in the second yellow arrow. Over 10 years into the future from when we drew our fibonacci extension, key .618 levels continue to hold importance in regard to support and resistance. 
There is also a tool called Fibonacci Retracement that is found from placing one level on a low and one level at the high. So, two points - This is often used when the level has seemingly finished it's rise from a low to a new high and to find out the key areas it will move down to during "consolidation" [Consolidation](https://www.investopedia.com/terms/c/consolidation.asp) 

The fibonacci extension in combination with the fibonacci retracement can define long and short term key levels. 
Note when drawing these, either go from the tip of the candlestick wicks to tip of candlestick wick, or the candlestick body to body, but that for longer term trends, combining a fibonacci retracement drawn from the wicks and a the same tool, a fibonacci retracement drawn from the bodies can outline "zones" - or entire areas you would be watching closely to maximize the trade (for example if you saw the volume of trades dying down, it would tip you off as to the correct timing. 

Also in the image is a diagonal channel. 
Looking at the monthly chart view of the S&P 500, after the 2008 crash, we can see the trend has been up. The technical definition of a Bear Market [BearMarket](https://www.investopedia.com/terms/b/bearmarket.asp) is when price falls by 20% of a peak for a period of time. That happened ~twice in the time frame we're analysing. When looking at this in the one month view of more than 10 years, things don't seem as alarming. In real life, the news is blasting panic and fear articles about the end of the market and the whole world crashing. If we we're using the channel as a reference point for what you were considering "red flags" then we would have mitigated losses during the relatively short crashes experienced in the past decade.

Based off the relative speed of transactions and market sentiment, the bot would enter into an incredibly short term trade (~1 second or much less) as to secure a sure micro profit in that moment. It can be useful in many ways and reading Real-time Transactions is called "Tape reading" and has an art to it. I could go into more detail some other time if you're interested in this aspect. 

The white box is for placing your orders. The options are Limit, Market, and Stop-Limit. Limits are for selecting the price you want to buy/sell at and amount. By placing a limit order, you are considered a Market Maker. You are setting the market prices. By using the Market Order, like mentioned before you are buying or selling at whatever is the available price. Buying/selling with a market order is considered being a "Market Taker". Generally speaking, using the Market order/being a market taker is considered bad. Then there is Stop-Limit, which places a market order only when the price moves to your specified limit. Other platforms have additional order types. For example OCO is One Cancels the Other, which could be useful for setting either a buy IF ___  or sell IF ___ order. An example of OCO being useful would be setting it overnight under the requirements that if the price didn't break past the important area you anticipated you could sell/buy. And conversely if it does break through you could buy/sell. 

The highlighted boxes as well as drawing trend lines and viewing different indicators are all built into these trading platforms. Interest Rates (*forex*)

The Market’s Driving Force.

Get to know the interest rate markets.

The entire global markets are inter-connected and spin on the gyrations in the interest rates. Each country we trade currencies with has their own interest rates and thus these rates influence the demand or disenchantment with it’s currency. The market will always seek yield and the higher yield the bigger the demand on average. There are several types of interest rates that traders can track and utilize to determine a trade idea. Understanding how they trade the yields will make long term trends easier to determine.



Interest rates spreads & differentials.

Track countries interest rates - just google historical spreadsheet data. The currency markets trend longer term based interest rates. Understanding which country has higher interest rates and which have lower rates can be a differential play or carry trade. Not only could you profit but interest is paid on top of that. If the country sees lowering rates, this will weaken the base currency... on the contrary, if the rates are increasing... this will prop up the currency.
